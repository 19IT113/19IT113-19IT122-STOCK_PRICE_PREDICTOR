```python
import pandas as pd
import statsmodels.api as sm
```

```python
data = pd.read_csv('RELIANCE.csv')
#Printing the dataset
print(data)
```

```python
data
```

```python
df = data['Adj_Close'].pct_change()*100
```

```python
df
```

```python
df = df.rename("Today")
```

```python
df
```

```python
df=df.reset_index()
```

```python
df
```

```python
for i in range(1,6):
    df['Lag'+str(i)]=df ['Today'].shift(i)
```

```python
df
```

```python
df['Adj_Volume']= data.Adj_Volume.shift(1).values/1000_000_000
```

```python
df = df.dropna()
```

```python
df
```

```python
df['Direction'] = [1 if i >0 else 0 for i in df['Today']] 
```

```python
df
```

```python
df = sm.add_constant(df)
```

```python
df
```

```python
X = df [['const','Lag1','Lag2','Lag3','Lag4','Lag5','Adj_Volume']]
```

```python
y = df.Direction
```

```python
model = sm.Logit(y,X)
```

```python
result=model.fit()
```

```python
result.summary()
```

```python
prediction = result.predict(X)
```

```python
prediction
```

```python
y
```

```python
def confusion_matrix(act,pred):
    predtrans = ['Up' if i > 0.5 else "Down" for  i in pred]
    actuals = ['Up' if i>0 else "Down" for i in act]
    confusion_matrix = pd.crosstab(pd.Series(actuals),
                                  pd.Series(predtrans),
                                   rownames=['Actuals'],
                                   colnames=['Predicted'])
    return confusion_matrix
```

```python
confusion_matrix(y,prediction)
```
